```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy
import requests
import pandas as pd
from sodapy import Socrata
import os

from collections import defaultdict


# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cdc.gov", None)

# Returned as JSON from API / converted to Python list of dictionaries by sodapy.
results = client.get_all("8xkx-amqh", date='2022-03-18')


state_to_vacci = defaultdict(list)
# Retrieve the column we want
for result in results:
  # print(result)
  item = {
            'date': result['date'],
            'recip_county': result['recip_county'],
            'recip_state': result['recip_state'],
            'completeness_pct': result.get('completeness_pct', None),
            'administered_dose1_recip': result.get('administered_dose1_recip', None),
            'series_complete_yes': result.get('series_complete_yes', None),
            'booster_doses': result.get('booster_doses', None)
          }
  
  # print(item)
  state_to_vacci[item['recip_state']].append(item)
states = state_to_vacci.keys()
print(state_to_vacci)

!mkdir vaccination_data_county
for state in states:
    state_vaccination_data = state_to_vacci[state]
    df = pd.DataFrame(columns=['date', 'recip_state', 'recip_county', 'completeness_pct', 'administered_dose1_recip', 'series_complete_yes', 'booster_doses'])
    for data in state_vaccination_data:
      df = df.append(data, ignore_index=True)
    p = os.path.join(os.getcwd(), "vaccination_data_county/vaccination_data_{}.csv".format(state))
    df.to_csv(p, index=False)
    display(df)
```
